```python
import pandas as pd
import numpy as np
```

```python
df7 = pd.read_csv('./df_security_concat_7.csv', encoding='cp949')
```

```python
df7
```

```python
df7_state_unique = df7.state.unique().tolist()
```

```python
df7_state_unique
```

```python
def top5(state_name):
    df_top5= df7[df7.state==state_name].sort_values(by='mean_hours' ,ascending=False).head(5)
    return df_top5
```

```python
top5_list = []
for state_name in df7_state_unique:
    top5_list.append(top5(state_name).new_state.tolist())
```
