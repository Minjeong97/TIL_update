#### 함수화

```python
def security_light(region):
    # 1. 파일 리스트 저장
    scl_lists = os.listdir(f'C:/Users/samsung/TIL/DS/project/miniPJ/01_pj/data/rawdata/rawdata_보안등/rawdata/{region}')
    # 2. 현재 작업 위치로 이동
    os.chdir(f'C:/Users/samsung/TIL/DS/project/miniPJ/01_pj/data/rawdata/rawdata_보안등/rawdata/{region}')
    # 3. 데이터 프레임 생성하고 모든 파일 합치기
    df = pd.DataFrame()
    for scl in scl_lists:
        df_raw = pd.read_csv(scl, encoding='cp949')
        df = pd.concat([df, df_raw])
    # 필요한 칼럼만 남기기
    df = df_raw[['설치개수','소재지도로명주소', '소재지지번주소', '위도', '경도', '데이터기준일자']]
    # 주소 칼럼 작업
    df['주소'] = np.where(pd.notnull(df['소재지도로명주소']) == True, df['소재지도로명주소'], df['소재지지번주소'])
    df = df.rename(columns={"주소": "address", "위도": "lat", "경도": "long", "설치개수": "install_cnt", "데이터기준일자": "date"})  # 칼럼명 변경
    # address -> state 제대로 되어 있는지 확인하고 'state'값 넣기
    df['address_check'] = np.where(df['address'].str.split(' ').str[0] == f'{region}', region, 'False')
    if df['address_check'] != 'False':
        df['state'] = df['address'].str.split().str[0]
        df['city'] = df['address'].str.split().str[1]
    # 칼럼 순서 바꾸기
    df = df[['address', 'state', 'city', 'lat', 'long', 'install_cnt', 'date']]
    
    return df
```

```python
# 함수 적용
regions = ['서울', '세종', '울산', '인천', '전라남도', '전라북도', '제주도', '충청남도', '충청북도']

# for region in regions:
#     security_light(region)
security_light('서울')
```
