```python
# Q1. chipotle 데이터를 로드해주세요.
```

```python
import pandas as pd

df_chipotle = pd.read_csv('./chipotle.tsv', engine='python', sep='\t') # hint sep 파라미터를 활용한다.
df_chipotle
```

```python
# Q2. 'quantity'의 값이 3인 데이터를 추출한다.
# 0부터 index를 재정렬한 후 상위 5개행만을 추출해보자.
```

```python
df_chipotle[df_chipotle.quantity == 3].reset_index().head()
```

```python
# Q3 'quantity', 'item_name', 'item_price'로만 구성된 데이터 프레임을
# new_df라고 정의해보자.
```

```python
new_df = df_chipotle[['quantity', 'item_name', 'item_price']]
new_df.head()
```

```python
# Q4 'item_name'이 'Chicken Burrito'인 데이터 프레임을 추출하고 인덱스를 초기화해보세요 ^^
```

```python
new_df[new_df.item_name == 'Chicken Burrito'].reset_index()
```

```python
# Q5 'item_name'중 'Burrito'가 포함된 경우의 데이터를 호출해주세요. 난이도-상
```

```python
new_df[new_df.item_name.str.contains('Burrito')]
```

```python
# Q6 'item_name'값이 Steak Salad 또는 Bowl인 데이터를 호출해주세요.- 난이도-중
```

```python
new_df[(new_df.item_name == 'Steak Salad') | (new_df.item_name == 'Bowl')]
```
