### DataFrame과 리스트, 딕셔너리, 넘파이 ndarray 상호변환

```python
import numpy as np
import pandas as pd
```

#### Data 생성

```python
# 3개의 컬럼으로 이름을 짓는다.
col_name1 = ['col1', 'col2', 'col3']

# (2,3)형태의 list를 생성 -> ndarray로 변형 -> DF로 변환
l2 = [[1,2,3],
    [11,22,33]]

array2d = np.array(l2)
print('array2 shape:\n', array2d.shape)

df_l2 = pd.DataFrame(l2, columns=col_name1)  # 2차원 데이터
print('2차원 리스트로 만든 df:\n', df_l2)

df_array2d = pd.DataFrame(array2d, columns=col_name1)
print('2차원 ndarray로 만든 df:\n', df_array2d)
```

```python
# dict -> dataframe 생성
dict123 = {
    'col1':[1,11],
    'col2':[2,22],
    'col3':[3,33],
}

df_dict123 = pd.DataFrame(dict123)
```

```python
df_dict123
```

#### DF를 다른 자료로 변환

```python
# 1. df.values를 통해 -> ndarray로 변환
array3_dic = df_dict123.values
print('\n array3_dic.values의 타입:', type(array3_dic), 
      '\n df_dict123.values의 shape:', array3_dic.shape)
```

```python
# 2. df.values -> (ndarray) -> list로 변환
df_dict123.values.tolist()

# 2. 참고
# DataFrame을 리스트로 변한
# list3 = df_dict.values.tolist()
# print('\n df_dict.tolist() 타입:', type(list3))
# print(list3)

# 3. df.to_dict() -> dict로 변환
df_dict123.to_dict('list')
```
